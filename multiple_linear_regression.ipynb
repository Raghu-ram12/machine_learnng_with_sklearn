```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("adagency.csv")
```

```python
df.head()
```

```python
x=df.iloc[:,:4]
```

```python
df.info()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(
    transformers=[("encoder", OneHotEncoder(), [3])], 
    remainder="passthrough",
    sparse_threshold=0  # Forces dense output
)
x = ct.fit_transform(x)
```

```python
x=pd.DataFrame(x)
```

```python
x
```

```python
x=x.drop(columns=0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=df["Profit"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,y_predict)
```

```python
metrics.mean_squared_error(y_test,y_predict)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,y_predict))
```

```python
# adjusted R value calculation
from sklearn.metrics import r2_score

from sklearn.metrics import r2_score

# Assuming y_test, y_predict, and X_test are defined
R = r2_score(y_test, y_predict) 
n = len(y_test)
p = x_test.shape[1] # p is the number of predictors (columns)
adj_r = 1 - (1 - R) * (n - 1) / (n - p - 1)
```
